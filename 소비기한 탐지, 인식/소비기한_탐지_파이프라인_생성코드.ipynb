# DMY Detection Running - 기존 논문 합성 데이터 running

#### 기존 논문 데이터에서 15000장만을 샘플링하여 base model 학습

```python
import os
import json
import random
import shutil
from tqdm import tqdm

# ====== 경로 설정 ======
base_dir = "/home/work/Seminar_Folder/A-team/세븐일레븐_사이드프로젝트/논문데이터/Date-Synth"
image_dir = os.path.join(base_dir, "images")
anno_path = os.path.join(base_dir, "annotations.json")

# ====== 출력 폴더 ======
output_img_dir = os.path.join(base_dir, "third/images_sampled")
output_label_dir = os.path.join(base_dir, "third/labels_sampled")
os.makedirs(output_img_dir, exist_ok=True)
os.makedirs(output_label_dir, exist_ok=True)

random.seed(42)

# ====== 이미지 목록 불러오기 ======
image_files = [f for f in os.listdir(image_dir) if f.lower().endswith((".jpg", ".png", ".jpeg"))]

# 15,000장 랜덤 선택
sample_size = min(15000, len(image_files))
sampled_images = random.sample(image_files, sample_size)

# ====== 주석 파일 로드 ======
with open(anno_path, "r", encoding="utf-8") as f:
    annotations = json.load(f)

# ====== 새 클래스 매핑 ======
# 0: 4글자 연도, 1: 영어 월, 2: 2자리 숫자
cls_map = {
    "Y4": 0,
    "ENG_MON": 1,
    "NUM2": 2,
}

def is_english_month(s: str) -> bool:
    if not s:
        return False
    s = s.strip().upper()
    return s in {
        "JAN", "FEB", "MAR", "APR", "MAY", "JUN",
        "JUL", "AUG", "SEP", "OCT", "NOV", "DEC"
    }

# ====== 변환 루프 ======
for img_name in tqdm(sampled_images, desc="Processing images"):
    src_img = os.path.join(image_dir, img_name)
    dst_img = os.path.join(output_img_dir, img_name)
    shutil.copy2(src_img, dst_img)

    if img_name not in annotations:
        continue

    img_info = annotations[img_name]
    width, height = img_info["width"], img_info["height"]

    txt_name = os.path.splitext(img_name)[0] + ".txt"
    txt_path = os.path.join(output_label_dir, txt_name)

    with open(txt_path, "w", encoding="utf-8") as f:
        for ann in img_info["ann"]:
            orig_cls = ann["cls"]          # 'day' / 'month' / 'year'
            trans = str(ann.get("transcription", "")).strip()

            # ===== 클래스 변환 규칙 =====
            new_cls = None

            if orig_cls == "year":
                # 4글자 -> Y4
                if len(trans) == 4 and trans.isdigit():
                    new_cls = "Y4"
                else:
                    # 2글자 연도 -> NUM2
                    new_cls = "NUM2"

            elif orig_cls == "day":
                # day는 전부 2자리 숫자 클래스로
                new_cls = "NUM2"

            elif orig_cls == "month":
                # month가 영어면 ENG_MON, 아니면 숫자로 처리
                if is_english_month(trans):
                    new_cls = "ENG_MON"
                else:
                    # 숫자형 month ('01','12'...) -> NUM2
                    new_cls = "NUM2"

            else:
                # 혹시 다른 거 있으면 스킵
                continue

            # 매핑 안 되면 스킵
            if new_cls not in cls_map:
                continue

            cls_id = cls_map[new_cls]
            x1, y1, x2, y2 = ann["bbox"]

            # YOLO 포맷 (정규화)
            x_center = ((x1 + x2) / 2) / width
            y_center = ((y1 + y2) / 2) / height
            w = (x2 - x1) / width
            h = (y2 - y1) / height

            f.write(f"{cls_id} {x_center:.6f} {y_center:.6f} {w:.6f} {h:.6f}\n")

print(f"✅ 완료! {sample_size}개의 이미지가 '{output_img_dir}'에 저장되었고,")
print(f"   해당 YOLO 라벨 파일이 '{output_label_dir}'에 생성되었습니다.")
```

```python
import os
import shutil
import random
from pathlib import Path

# ====== 설정 ======
img_dir = Path("/home/work/Seminar_Folder/A-team/세븐일레븐_사이드프로젝트/논문데이터/Date-Synth/third/images_sampled")
txt_dir = Path("/home/work/Seminar_Folder/A-team/세븐일레븐_사이드프로젝트/논문데이터/Date-Synth/third/labels_sampled")
base_out = Path("/home/work/Seminar_Folder/A-team/yolo_dmy_label_third_dataset")
random.seed(42)

# YOLO 폴더 구조 생성
for sub in ["images/train", "images/val", "labels/train", "labels/val"]:
    os.makedirs(base_out / sub, exist_ok=True)

# 이미지 확장자
valid_exts = {".jpg", ".jpeg", ".png", ".JPG", ".PNG"}

# ====== 이미지-라벨 매칭 ======
pairs = []
for img_file in img_dir.iterdir():
    if img_file.suffix in valid_exts:
        label_file = txt_dir / (img_file.stem + ".txt")
        if label_file.exists():
            pairs.append((img_file, label_file))

print(f"✅ 총 {len(pairs)}개의 이미지-라벨 쌍이 발견되었습니다.")

# ====== 9:1 분할 ======
random.shuffle(pairs)
split_idx = int(len(pairs) * 0.9)
train_pairs = pairs[:split_idx]
val_pairs = pairs[split_idx:]

# ====== 복사 함수 ======
def copy_pairs(pairs, subset):
    for img, txt in pairs:
        shutil.copy2(img, base_out / f"images/{subset}/{img.name}")
        shutil.copy2(txt, base_out / f"labels/{subset}/{txt.name}")

copy_pairs(train_pairs, "train")
copy_pairs(val_pairs, "val")

print(f"📦 Train: {len(train_pairs)}개, Val: {len(val_pairs)}개 복사 완료!")

# ====== data.yaml 생성 ======
yaml_path = base_out / "data.yaml"

# ⚙️ 전체 클래스 목록
all_classes = ["Y4", "ENG_MON",  "NUM2"]

# ⚙️ 제외할 클래스
ignore_classes = {}

# ⚙️ 필터링된 클래스 목록
used_classes = [c for c in all_classes if c not in ignore_classes]

yaml_content = f"""train: {base_out / 'images/train'}
val: {base_out / 'images/val'}

nc: {len(used_classes)}  # 클래스 개수
names: {used_classes}
"""

with open(yaml_path, "w", encoding="utf-8") as f:
    f.write(yaml_content)

print(f"✅ data.yaml 생성 완료: {yaml_path}")
print(f"🗂️ 포함된 클래스: {used_classes}")
print("🎉 YOLO 학습 준비 완료!")
```
